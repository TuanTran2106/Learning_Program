```python
df2 = df.melt(id_vars=["Code", "Description"], 
        var_name="Date", 
        value_name="Value")
df2
```

```python
df3 = df2[df2['Date'].apply(lambda x: x.endswith(".1"))]
df3
```

```python
df3['Date'] = df3['Date'].apply(lambda x: x.replace(".1", ''))
df3
```

```python
df4 = df2[~df2['Date'].apply(lambda x: x.endswith(".1"))]
df4
```

```python
df5 = pd.merge(df3, df4, on=["Code",'Date'])
df5
```

```python
df5.columns = ['Code', 'Description', 'Date', 'Sales Value (Incl GST)', 'xx', 'Units Sales']
```
